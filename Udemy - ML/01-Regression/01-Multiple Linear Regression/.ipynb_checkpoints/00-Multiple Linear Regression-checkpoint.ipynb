## Assumptions of Linear Regression

```python
import subprocess
path = 'C:\\Users\\Dell\\Documents\\GitHub\\Tutorials-Studying\\Udemy - ML\\01-Regression\\SuperDataScience-Poster-Assumptions-of-Linear-Regression.pdf'
subprocess.Popen([path], shell=True)
```

___
1. __Linearity__
<br>We want to make sure that there is a linear relationship between our dependent variable
and each independent variable.
___
2. __Homoscedasticity (equal variance)__
<br>Meaning that you don't want to see a cone type shape on your chart whether an increasing cone
or a decreasing cone.
Which would mean that variance is dependent on the independent variable.
___

3. __Normality of Error Distribution (Multivariate Normality)__
___
4. __No Autocorrelation (Independence of Observations)__
<br>What that means is that we don't want to see any kind of pattern in our data. A pattern in the data
indicates that our rows are not independent,
that some rows are affecting other rows
and other rows, etc. A classic example of this would be the stock market
where previous prices affect future prices,
which affect future prices and so on.
___
5. __Lack of Multicollinearity (Predictors are NOT correlated with each other)__
<br>Basically, we want our independent variables or predictors not to be correlated with each other.
<br>If they're not correlated,
then we can build a linear regression.
<br>If they are correlated,
then if we do proceed and build a linear regression model,
then the coefficient estimates that we get in the model
will become unreliable.
___
6. __The Outlier Check__ (this is not an assumption, just an __'extra' check__)
So something that we want to consider is, should we remove the outliers before building a linear regression or do we want to build a linear regression with the outliers included?
This will depend on your business knowledge
and knowledge of the data set.

The approach that you need to take when you face __categorical variables__ in regression models is you need to __create dummy variables__.

- First you need to go through your column and find all the different (distinct) categories you have.
- For every single category that you followed you need to create a new column.
- Add respectively 0 and 1.

```python
Image.open('dummy.jpg')
```

In this case we have two categories - New York and California.

__!!!__
__You don't use the California column when building your regression model__.
<br>All the information in our data is preserved even if we just stick to the one New York column because you can tell right away if one is a one then it's a company that works and it operates in New York.
If one is a zero it's a company that operates in California.

Basically here we have a coefficient in our equation for New York, __but for California
there's no coefficient__ because when D1 is zero that whole last part of the equation becomes zero and might seem biased at first but basically what that means is that the coefficient for California __is__ going to be __included in the constant b0__.

## Dummy Variable Trap

What will happen if we include the second dummy variable in the model as well?

y = b0 + b1 * x1 + b2 * x2 + b3 * x3 + b4 * D1 + b5 * D2 

You're basically duplicating a variable, as __D1 = 1 - D2__
<br>The phenomenon where one or several independent variables in a linear regression predict another is called __multicollinearity__. <br>As a result of this effect the model __cannot distinguish between the effects of D1 from the effects from of D2__. <br>And therefore it won't work properly. And this is called the __dummy variable trap__.

## !!!

To sum up whenever you're building a model always __omit one dummy variable__.

## Hypothesis testing, P-Value and Statistical Significance

Is my results statistically significant or not? Or are these insights statistically significant or not?
<br>When we're talking about hypothesis testing,
we're thinking about __two possible alternative universes__.

H0: This is a fair coin. (__null hypothesis__)
<br>H1: This isn't a fair coin. (__alternative hypothesis__)

We're trying __to assess which one is True__.

- So the way we're gonna go about this is we're going to assume that H0 is true.
- And then based on our experiment, we're going to see if we can contradict that.
- So statistical significance is basically usually 5%. I am so unlikely to see this by random that I'm going to reject this null hypothesis.
- And you can set this confidence level to whatever you like, you can set it to 1%, 10%,.. So it depends on your experiment (for example in medical experiments usually 1%)

# Multiple Linear Regression Implementation in Python

## 1. Data Preprocessing

#### Importing the libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

#### Importing the dataset

```python
df = pd.read_csv('50_Startups.csv')
```

```python
df.head()
```

#### Checking for missing data

```python
df.isnull().sum()

# we don't have any missing values
```

```python
X = df.iloc[:, :-1].values
y = df.iloc[:, -1].values
```

#### Encoding categorical data - independent variable

```python
# encode 'State' column
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
```

```python
ct = ColumnTransformer(transformers=[('encoder', OneHotEncoder(), [3])], remainder='passthrough')
X = ct.fit_transform(X)
```

```python
X

# three new columns at the beginning, encoding that 'State' variable
```

```python
y
```

## !!!

__This is important__:

__We don't have to apply feature scaling in case of Multiple Linear Regression.__ 
<br>Why? 
- Because in the equation of the Multiple Linear Regression, you have beta coefficient that is multiplied to each independent variable, each feature. And, therefore, it doesn't matter that some features have higher values than others, because the __coefficients will compensate to put everything on the same scale__.

- And, therefore, remember this, in Multiple Linear Regression, there is __absolutely NO NEED to apply feature scaling__.

---

## !!!

I know a lot of you asked this question.
<br>__Do we need to check the assumptions of linear regression?__

- The answer is __absolutely NOT__, because whenever you have a new data set, and you want to experiment with some ML models to figure out which one leads to the highest accuracy, even if your data set doesn't have linear relationships, you can still try a Multiple Linear Regression on it.
- And, __if__ you know your __data set doesn't have linear relationships__, well, your __Multiple Linear Regression model__ will just __perform poorly__, and therefore, it will get __an accuracy lower__ than the accuracies of your other models. 
- So, you will just not select the Multiple Linear Regression model, but you don't have to check the Multiple Linear Regression assumptions. It will just be a waste of time.

---

#### Splitting the dataset into Training Set and Test Set

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

---

## !!!

I would like to answer two important questions.

- The first question is, __do we have to do something to avoid the dummy variable trap__?
    - And the answer is __NO__, because indeed the Multiple Linear Regression
class will automatically avoid this trap which means that one of the three first columns here will automatically be outcast ().
__You don't have to remove one of the columns here.__

- second question, __do we have to work on the features to select the best ones with the techniques of building models__, like, for example, backward elimination.
    - And the answer is once again, __NO__. Why is that? Well, for the exact same reason as the dummy variable trap. The class that we're about to call to build our Multiple Linear Regression model will automatically identify the best features, the features that are the most statistically significant to figure out how to predict the dependent variable.

---

## 2.1. Training the Multiple Linear Regression Model on the Training set

The good news is actually that the class that we're about to use to build and train this Multiple Linear Regression model
is actually the exact same class as for the Simple Linear Regression model, it will just recognize that there are several features and therefore that we are doing multiple linear regression.

```python
from sklearn.linear_model import LinearRegression
```

```python
regressor = LinearRegression()
regressor.fit(X_train, y_train)
```

## 2.2. Predicting the Test set results

We actually have four features instead of one and therefore we cannot actually plot a graph.

```python
y_pred = regressor.predict(X_test)
```
