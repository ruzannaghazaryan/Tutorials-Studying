A groupby() operation allows us to examine data on a __per category__ basis.

We need to choose a __categorical__ column to call with __groupby()__.

__Keep in mind__ that __categorical columns__ can actually still be __numerical__.
<br> (For example, the cabin class category on a ship or a plane like first class, second class or third class, those could just be represented as numbers one, two and three.
<br>Another example could be the model year for a dataset consisting of cars. You can have the category of 1970, 1971, 1972 and so on, just different years.)

___

__!! Note__ that in pandas calling groupby() by itself creates a _'lazy'_ groupby object waiting to be evaluated by an aggregate method call.

#### Adding an aggregate method call. To use a grouped object, you need to tell pandas how you want to aggregate the data.

Common Options:

    mean(): Compute mean of groups
    sum(): Compute sum of group values
    size(): Compute group sizes
    count(): Compute count of group
    std(): Standard deviation of groups
    var(): Compute variance of groups
    sem(): Standard error of the mean of groups
    describe(): Generates descriptive statistics
    first(): Compute first of group values
    last(): Compute last of group values
    nth() : Take nth value, or a subset if n is a list
    min(): Compute min of group values
    max(): Compute max of group values
    
Full List at the Online Documentation: https://pandas.pydata.org/docs/reference/groupby.html

<a id='groupby_one_column'></a>
__`.groupby() method by one column`__

```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('mpg.csv')
```

```python
df.head()
```

___

Start exploring how maybe performance has changed for the cars throughout the years.

```python
df['model_year'].unique()
```

```python
df['model_year'].value_counts()
```

Trying to answer a question like how has MPG (miles per gallon) changed on average per year?

```python
df.groupby('model_year')
```

```python
df.groupby('model_year').mean()
```

So what it's going to do, it is going to grab the column you want, to group by that particular category, set that as the index, and then it takes in the aggregate of what it can do.



```python
# and if you're interested in certain column

df.groupby('model_year').mean()['mpg']
```

<a id='groupby_multiple'></a>
__`.groupby() method multiple columns`__

<a id='#groupby_multiple'></a>
<u> multilevel or hierarchy index

```python
df.groupby(['model_year', 'cylinders']).mean()
```

```python
df.groupby(['model_year', 'cylinders']).mean().columns
```

```python
df.groupby(['model_year', 'cylinders']).mean().index
```

```python
df.groupby(['model_year', 'cylinders']).mean()['mpg']
```

```python
df.groupby('model_year').describe().transpose()
```

```python
year_cyl = df.groupby(['model_year', 'cylinders']).mean()
year_cyl
```

```python
year_cyl.index.names
```

```python
year_cyl.index.levels
```

```python
# grabbing by outside level index

year_cyl.loc[70]
```

```python
year_cyl.loc[[70, 82]]
```

```python
year_cyl.loc[(70, 8)]
```

___

<a id='cross_section'></a>

__`.xs() method`__ 

a cross_section, which is a more general way of grabbing something that has multilevel index.

```python
year_cyl
```

```python
year_cyl.xs(key=70, level='model_year')
```

Note that we could only put one value as the key. If you want more than one value, you should use .loc[[]]

```python
year_cyl.xs(key=[70, 80], level='model_year')
```

```python
year_cyl.loc[[70, 80]]
```

___

Let's imagine I want to grab only the four cylinder cars, but I wanted to do it for every single year.

```python
year_cyl.xs(key=4, level='cylinders')
```

__!!!__ And let's say I actually want the cylinders for six and eight for every single year, well, that leaves us in a bit of a conundrum (), because I just told you we could only put one value as the key.

And if you're trying to select certain cylinders, then it's probably much easier to filter out those values before you run the groupby.

<a id='isin'></a>
__`.isin() method`__ 

```python
df[df['cylinders'].isin([6, 8])].groupby(['model_year', 'cylinders']).mean()
```

So sometimes it's a lot easier to do the filtering and conditional filtering on the data frame first and then do your aggregate functions.

<a id='swaplevel'></a>
__`.swaplevel() method`__

If for some reason you want to swap these, make 'cylinders' the outer level and 'model_year' the inner level, then

```python
year_cyl.swaplevel()
```

```python
year_cyl.swaplevel().loc[4]
```

<a id='agg'></a>

__`.agg() method`__

So what I want to show you is how you can actually specify a particular aggregate function or group of aggregate functions for a particular column.

```python
df.agg(['std', 'mean'])
```

```python
df.agg(
    {
    'mpg': ['max', 'mean'],
    'weight': ['mean', 'std']
    }
)
```
