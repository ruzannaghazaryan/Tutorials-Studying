```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

### adding legends

legend (explanation) - the words written on or next to a picture, map, coin, etc. that explain what it is about or what the symbols on it mean.


```python
x = np.linspace(0, 11, 10)
```

<a id='id1'></a>

```python
fig = plt.figure()
ax = fig.add_axes([0, 0, 1, 1])
ax.plot(x, x)

# plot again on the exact same axes
ax.plot(x, x ** 2)
```

<a id='id2'></a>
Adding in the legends that will tell what each color represents

```python
fig = plt.figure()

ax = fig.add_axes([0, 0, 1, 1])

ax.plot(x, x, label='X vs X')
ax.plot(x, x ** 2, label='X vs X^2')

ax.legend()
```

___

<a id='id3'></a>
You can actually tell matplotlib __where to place the legend__.

```python
fig = plt.figure()

ax = fig.add_axes([0, 0, 1, 1])

ax.plot(x, x, label='X vs X')
ax.plot(x, x ** 2, label='X vs X^2')

ax.legend(loc='lower right')
```

Or to manually set the location pass in a tuple, which corresponds with the exact same two parameters here for the bottom left hand corner.

```python
fig = plt.figure()

ax = fig.add_axes([0, 0, 1, 1])

ax.plot(x, x, label='X vs X')
ax.plot(x, x ** 2, label='X vs X ** 2')

ax.legend(loc=(1.1, .5))
```

```python
fig = plt.figure()

ax = fig.add_axes([0, 0, 1, 1])

ax.plot(x, x, label='X vs X')
ax.plot(x, x ** 2, label='X vs X ** 2')

ax.legend(loc=(.5, .5))
```

```python
fig = plt.figure()

ax = fig.add_axes([0, 0, 1, 1])

ax.plot(x, x, label='X vs X')
ax.plot(x, x ** 2, label='X vs X ** 2')

ax.legend(loc=(-.5, .5))
```

### line styles

<a id='id5'></a>
- __linewidth or lw parameter__

```python
fig = plt.figure()

ax = fig.add_axes([0, 0, 1, 1])

ax.plot(x, x, color='#1df5cd', lw=10)
ax.plot(x, x+1, color='purple', linewidth=.4)
```

___

<a id='id6'></a>
- __ls or linestyle parameter__

```python
fig = plt.figure()

ax = fig.add_axes([0, 0, 1, 1])

ax.plot(x, x, color='#1df5cd', lw=10, ls='--')
ax.plot(x, x+1, color='purple', lw=2, linestyle='-.')
```

___

Or you can also create your own custom styling with .set_dashes() command.

```python
fig = plt.figure()

ax = fig.add_axes([0, 0, 1, 1])

lines = ax.plot(x, x+1, color='purple', lw=5)
```

```python
lines
```

```python
type(lines)
```

So we need to say lines[0] in order to grab the line object from the list.
<br>The way .set_dashes() command works is you pass in a list that corresponds with your custom dash sequence.

```python
fig = plt.figure()

ax = fig.add_axes([0, 0, 1, 1])

lines = ax.plot(x, x+1, color='purple', linewidth=5)

lines[0].set_dashes([1, 1, 1, 2, 3, 5])
```

```python
fig = plt.figure()

ax = fig.add_axes([0, 0, 1, 1])

lines = ax.plot(x, x+1, color='purple', lw=5)

lines[0].set_dashes([5, 2, 10, 2])

# five solid points, then two blank points, then ten solid points and then two blank points
```

So the way this works is you essentially say how many points you want and then how many points you want spaced after it.

### markers

There are string codes for the marker type.

```python
fig = plt.figure()

ax = fig.add_axes([0, 0, 1, 1])

ax.plot(x, x+1, color='purple', lw=3)
```

This is actually not truly a continuous line. In fact, if we look at what we're actually plotting - x, it's just some points. So let's imagine they actually wanted to plot each of those points that's actually occurring here.

```python
fig = plt.figure()

ax = fig.add_axes([0, 0, 1, 1])

ax.plot(x, x+1, color='purple', ls='--', lw=2, marker='o')
```

__markersize or ms parameter__

```python
fig = plt.figure()

ax = fig.add_axes([0, 0, 1, 1])

ax.plot(x, x+1, color='purple', ls='--', lw=2, marker='o', ms=10)
ax.plot(x, x, marker='+', markersize=20)
```

___

__markerfacecolor, markeredgewidth, markeredgecolor parameters__

```python
fig = plt.figure()

ax = fig.add_axes([0, 0, 1, 1])

ax.plot(x, x+1, color='purple', lw=2, marker='o', ls='--', ms=10,
       markerfacecolor='red', markeredgewidth=8, markeredgecolor='orange')
```

```python
fig = plt.figure()

ax = fig.add_axes([0, 0, 1, 1])

ax.plot(x, x+1, color='purple', lw=2, marker='o', ls='--', ms=10,
       markerfacecolor='red', markeredgewidth=4, markeredgecolor='orange')
```

```python
fig = plt.figure()

ax = fig.add_axes([0, 0, 1, 1])

ax.plot(x, x+1, color='purple', lw=2, marker='s', ls='--', ms=10,
       markerfacecolor='red', markeredgewidth=4, markeredgecolor='orange')
```
