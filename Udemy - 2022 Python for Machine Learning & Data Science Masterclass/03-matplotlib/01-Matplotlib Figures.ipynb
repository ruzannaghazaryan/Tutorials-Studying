- The more comprehensive Matplotlib OOP (object oriented programming) API makes use of a Figure object. (__a Figure object as a blank canvas__) This means we will instantiate figure objects and then call methods or attributes from that object.
- We then add axes to this Figure object
- and then plot on those axes.

```python
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
from PIL import Image
```

### creating a figure

```python
plt.figure(); # blank canvas
```

<a id='id2'></a>

```python
plt.figure(figsize=(10, 10));
```

<a id='id3'></a>

```python
# the next step is to add a set of axes to actually plot on

fig = plt.figure()
axes = fig.add_axes([0, 0, 1, 1])
```

- The first two parameters in this set of axes call are the lower left corner of the axes. And that is all in relationship to the underlying figure, which means if you add a set of axes with zero zero as its lower left corner, that means zero zero in regards to the figure. So the lower left corner of the axes is the same as the lower left corner of the underlying figure canvas.
- The second set of parameters is the width and height of the axes. Passing in 1, 1 means take up 100 percent above the width and height of the underlying figure.
- So pretty much all these numbers are in ratios in regards to the underlying figure.

### adding multiple set of axes to the same figure

```python
type(fig)
```

I can add as many set of axes as I want, and then I can plot on each of those axes.

```python
fig = plt.figure()

# Large axes
axes1 = fig.add_axes([0, 0, 1, 1])
axes1.plot(a, b)

# Small axes
axes2 = fig.add_axes([0.2, 0.2, 0.5, 0.5])
axes2.plot(x, y);
```

This probably isn't super useful to have an insert, which is a completely different plot.

```python
fig = plt.figure()

# Large axes
axes1 = fig.add_axes([0, 0, 1, 1])
axes1.plot(a, b)
axes1.set_xlim(0, 8)
axes1.set_ylim(0, 8000)
axes1.set_xlabel('A')
axes1.set_ylabel('B')
axes1.set_title('power of 4')


# Small axes
axes2 = fig.add_axes([0.2, 0.2, .3, .3])
axes2.plot(a, b)
axes2.set_xlim(1, 2)
axes2.set_ylim(0, 55)
axes2.set_xlabel('A')
axes2.set_ylabel('B')
axes2.set_title('Zoomed in')
```

### figsize and dpi parameters of plt.figure()

```python
fig = plt.figure()

axes1 = fig.add_axes([0, 0, 1, 1])
axes1.plot(a, b)
```

You'll notice that it looks a little blurry (размытым) here and you can fix that by editing the dpi (the resolution of the figure in dots-per-inch).

```python
fig = plt.figure(dpi=100)

axes1 = fig.add_axes([0, 0, 1, 1])
axes1.plot(a, b)
```

It starts to get more and more clear. But don't set it too high, otherwise it'll take up a lot of RAM to render something with such high resolution on your notebook.

___

```python
fig = plt.figure(figsize=(6, 3), dpi=100)
axes = fig.add_axes([0, 0, 1, 1])
axes.plot(a, b)
```

### saving the figure

```python
fig = plt.figure(figsize=(2, 2))
axes = fig.add_axes([0, 0, 1, 1])
axes.plot(a, b);
fig.savefig('new_image.png')
```

<a id='id8'></a>
To include the X and Y axes in the saved image and their numbers use __bbox_inches__ parameter (bounding box).

```python
fig = plt.figure(figsize=(2, 2))
axes = fig.add_axes([0, 0, 1, 1])
axes.plot(a, b);
fig.savefig('new_image_with_axes.png', bbox_inches='tight')
```
