## Functional Programming Method for Matplotlib

__!!!__ If you are running __.py scripts__ instead of .ipynb notebooks, you will need to add the __plt.show()__ command.

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

You need to run a "magic" command to see the plots inline with the notebook.

```python
import numpy as np
```

```python
x = np.arange(0, 10)
```

```python
y = x * 2 

# linear relationship (y = kx + b)
```

```python
x
```

```python
y
```

```python
plt.plot(x, y)

# a linear plot between the coordinates
```

```python
plt.plot(x, y)
plt.show()
```

As soon as I added plt.show(), I actually didn't see a little extra output, which is [<matplotlib.lines.Line2D at 0x23347478c08>].
<br>Jupyter Notebook shows you both the actual object that's being returned at what location in your computer's memory and the rendering of the visualization itself.
<br>So in order to only see the rendering of the visualization, you can 
- either add plt.show()
- or, for jupyter notebook users, they can add a semicolon, which basically says, just show me the visualization itself.

```python
plt.plot(x, y);
```

___

```python
plt.plot(x, y)
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.title('String Title');
```

__!!! A really important thing to note here__ is if you're using Jupyter notebook, these should all go within the same cell. Otherwise, you'll get errors because matplotlib won't be able to connect the label to the plot.

___

We can see a little bit of negative along y axis and a little bit of negative below zero along the x axis.

```python
plt.plot(x, y)
plt.title('String Title')
plt.xlabel('x axis') 
plt.ylabel('y axis')
plt.xlim(0, 6); # Lower Limit, Upper Limit
```

```python
plt.plot(x, y)
plt.title('String Title')
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.xlim(0, 6)
plt.ylim(0, 15); # Lower Limit, Upper Limit
```
