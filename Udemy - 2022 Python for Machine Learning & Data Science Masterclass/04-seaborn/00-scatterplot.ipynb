- Scatterplots show the relationship between two continuous features.
- Recall that __continuous features__ are numeric variables that can take any number of values between any two values.
- __Categorical features__ represent distinct and unique categories (where there's no real kind of in-between value sometimes).

___

Continuous Feature Examples
- Age 
- Height
- Temperature
- Salary
- Prices

Categorical Feature Examples
- Shapes
- Names
- Colors

```python
import seaborn as sns
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv('dm_office_sales.csv')
```

```python
df.head()
```

___

Scatterplot _ each X and Y coordinate point gets a dot. So each of those dots represents a particular employee.

If someone asked us what is the relationship between salary and sales,
- we can try to answer that numerically. Maybe we take the Pearson correlation coefficient between salary and sales, see if they're correlated.
- or what might be nicer is to just visualize this. So that's where the scatterplot comes in.

___

<a id='xydata'></a>
__`sns.scatterplot(x='', y='', data=)`__

```python
sns.scatterplot(x='salary', y='sales', data=df)

# So clearly here we can see a trend where the more sales you make, the greater your salary is going to be.
```

__!!! If running this as a .py script, don't forget to add plt.show()__

___

seaborn is actually creating this with matplotlib, and that means you can actually grab this figure canvas and start editing it before you even call scatterplot with seaborn.

Let's imagine I wanted to edit this figure size.

```python
plt.figure(figsize=(12, 4))
sns.scatterplot(x='salary', y='sales', data=df)
```

seaborn is just calling matplotlib on the backend, which means if you start defining a figure right before you call seaborn, seaborn is actually going to then adjust to those figure parameters.

```python
# if you want the images to be a little clearer, you can also up the dpi

plt.figure(figsize=(12, 4), dpi=200)
sns.scatterplot(x='salary', y='sales', data=df)
```

___

### seaborn parameters

<a id='hue'></a>
__`sns.scatterplot(hue=)`__

And essentially it's going to ask for what should you color these dots by? If you don't provide a hue, it just colors them all the same value.

```python
sns.scatterplot(x='salary', y='sales', data=df, hue='level of education')
```

It automatically creates a legend for you and colors the points based off this particular category.

__Notice that__ 'level of education' was a __categorical column__.

___

Now let's choose a __continuous column__, for example 'salary', and see what happens.

```python
sns.scatterplot(x='salary', y='sales', data=df, hue='salary')
```

What it's going to do is it's going to essentially create a gradient for you.

I am now showing on a two dimensional plot three dimensions of information, the salary is one dimension, sales is the second dimension, and then a hue parameter is technically a third dimension.

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='salary', y='sales', data=df, hue='work experience')
```

___

<a id='palette'></a>

__`sns.scatterplot(palette=)`__

How do I get to choose what colours are here? 
<br>The way that works is you have to choose a palette from matplotlib's color map.

https://matplotlib.org/stable/tutorials/colors/colormaps.html

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='salary', y='sales', data=df,
               hue='level of education',
               palette='Set3')
```

___

<a id='size'></a>
__`sns.scatterplot(size=)`__

What I could do is say, I want the size of the points to be based off 'salary' as well, just to make it even more readable.

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='salary', y='sales', data=df, size='salary')

# the bigger your salary is, the bigger this point is
```

___

<a id='s'></a>

__`sns.scatterplot(s=)`__

What if I want to make all the markers a little larger?

```python
sns.scatterplot(x='salary', y='sales', data=df, s=200)
```

___

<a id='alpha'></a>

__`sns.scatterplot(alpha=)`__

alpha parameter just means transparency

In the above plot you're inevitably going to get some points stacked right on top of another point. In such cases you can set an alpha parameter.

```python
sns.scatterplot(x='salary', y='sales', data=df, s=80, alpha=0.2)
```

___

<a id='style'></a>
__`sns.scatterplot(style=)`__

What it's going to do is instead of having everything be a dot, it's going to automatically shape things based off the style.

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='salary', y='sales', data=df, style='level of education', s=120)
```

And then if you want, you can always combine this with a hue parameter.

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='salary', y='sales', data=df, s=120,
               hue='level of education',
               style='level of education')
```

__Keep in mind__ that style should be a categorical feature because you're categorizing things into different shape.

___

<a id='save'></a>

### saving the plot

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='salary', y='sales', data=df, 
               hue='level of education', 
               style='level of education',
               s=120)

plt.savefig('my_plot.jpg')
```
