---

- __sns.jointplot()__
      - We can map histograms to each feature of a scatterplot to clarify the distributions within each feature.
      - We can also adjust the scatterplot to be a hex plot or a 2D KDE plot... 
- __sns.pairplot()__
      - It's a quick way to compare all numerical columns in a DataFrame.
      - It automatically creates a histogram for each column and a scatterplot comparison between all possible combinations of columns.
      - Keep in mind that this pairplot() operation is only going to operate on numerical columns.
__<u>Warning!!!</u>__

    - pairplot() can be CPU and RAM intensive for large DataFrames with many columns.
    - so it's a good idea to first filter down to only the columns that you're interested in before calling on the full DataFrame.

## Jointplot

So what we need here are two continuous features, one for the x axis and then one for the y axis. And we have these continuous features - math score, reading score and writing score, essentially allowing us to answer the question, is there a relationship between the math score and the reading score or math score and the writing score, if someone scores really well in math, are they also really likely to score well on reading and writing?

What you see here is the scatterplot 'math score versus reading score' with the additional information of the distribution of each of those features.

__`sns.jointplot(data=, x='', y='')`__

```python
sns.jointplot(data=df,
             x= 'math score',
             y='reading score')
```

___

__hexagonal plot__ inside instead of a scatterplot

It counts how many points fall within each hexagon and then just darken the hexagons based on how many points fall within each one (darker areas - a lot of points stacked). And this is nice because I'm not entirely sure how many points are stacked on top of each other in case of a scatterplot.

__`sns.jointplot(data=, x='', y='', kind='hex')`__

```python
sns.jointplot(data=df,
             x='math score',
             y='reading score',
             kind='hex')
```

__`sns.jointplot(data=, x='', y='', alpha=)`__

```python
sns.jointplot(data=df,
             x='math score',
             y='reading score',
             alpha=0.2)
```

___

What's nice about this is it lines up with the bins of the histograms themselves.

__`sns.jointplot(data=, x='', y='', kind='hist')`__

```python
sns.jointplot(data=df,
             x='math score',
             y='reading score',
             kind='hist')
```

___

__`sns.jointplot(data=, x='', y='', kind='kde')`__

```python
sns.jointplot(data=df,
             x='math score',
             y='reading score',
             kind='kde')
```

__`sns.jointplot(data=, x='', y='', kind='kde', shade=True)`__

```python
sns.jointplot(data=df,
             x='math score',
             y='reading score',
             kind='kde',
             shade=True)
```

___

__`sns.jointplot(data=, x='', y='', hue='')`__

```python
sns.jointplot(data=df,
             x='math score',
             y='reading score',
             hue='gender')
```

## Pairplot

- We always want to be careful with it because it is quite CPU intensive.
- Along the diagonal, we have the histograms for each of those continuous features.
- as well as the relationship between each of combinations, and because of this, you essentially get duplicates along the diagonal.

__`sns.pairplot(data=)`__

```python
sns.pairplot(data=df)
```

---

When you're separating out different categories (hue parameter), it's really hard to read in a stacked histogram or histogram stacked on top of each other, which is why when you add in that hue, it automatically fixes the diagonal to be a KDE for you, which is much easier to read when you're comparing two separate categories.

__`sns.pairplot(data=, hue='')`__

```python
sns.pairplot(data=df, hue='gender')
```

___

__`sns.pairplot(data=, hue='', palette='')`__

```python
sns.pairplot(data=df, hue='gender', palette='viridis')
```

---

However, it should be noted, you can always specify that you do want the diagonal to remain a histogram.

__`sns.pairplot(data=, hue='', diag_kind='hist')`__

```python
sns.pairplot(data=df, hue='gender', diag_kind='hist')
```

---

To get rid of the duplicates and just show one corner:

__`sns.pairplot(data=, corner=True)`__

```python
sns.pairplot(data=df, hue='gender', corner=True)
```
