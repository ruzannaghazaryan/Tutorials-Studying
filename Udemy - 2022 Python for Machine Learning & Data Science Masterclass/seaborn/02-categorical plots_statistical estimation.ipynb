___

- The categorical plots are going to focus on displaying a statistical metrics __per__ category. Basically this means we will visually report back some statistic (such as mean or count) in a plot.
- It is the visualization __equivalent of a groupby() call__.

___

The two main types of plots for this are:
- __countplot()__
    - Counts number of rows per category.
    
    
- __barplot()__
    - General form of displaying any chosen metric per category.

___

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

import seaborn as sns
```

```python
df = pd.read_csv('dm_office_sales.csv')
```

```python
df.head()
```

## Countplot

<a id='valuecount'></a>

__`.value_counts() method`__

```python
df['division'].value_counts()
```

<a id='plot1'></a>

__`sns.countplot(data=, x='')`__

```python
sns.countplot(data=df, x='division')
```

There's some crossover here, so let's make the plot a little bigger.

```python
plt.figure(figsize=(9, 4), dpi=80)
sns.countplot(data=df, x='division')
```

```python
df['level of education'].value_counts()
```

```python
plt.figure(figsize=(9, 4))
sns.countplot(data=df, x='level of education')
```

___

<a id='plot2'></a>
__Keep in mind,__ we could add a `hue parameter` to split these up into further categories.

__`sns.countplot(data=, x='', hue='')`__

```python
plt.figure(figsize=(9, 4))
sns.countplot(data=df, x='level of education', hue='division')
```

___

<a id='plot3'></a>
You can also choose coloring - `palette parameter`
https://matplotlib.org/3.1.1/gallery/color/colormap_reference.html

__`sns.countplot(data=, x='', hue='', palette='')`__

```python
plt.figure(figsize=(9, 4))
sns.countplot(data=df, x='level of education', hue='division', palette='Set2')
```

## Barplot

<a id='plot4'></a>
What I have to decide is what I'm actually going to show as the aggregation method per group. It is the `estimator parameter` where you pass in the numpy function that you want to call.

__`sns.barplot(data=, x='', y='', estimator=, ci=)`__

```python
# calculate the mean salary per level of education

plt.figure(figsize=(9, 5), dpi=80)
sns.barplot(data=df, x='level of education', y='salary', estimator=np.mean, ci='sd')
```

The black line on top of that bar represents the standard deviation (ci='sd'). 
<br>Visually, I can see that the master's degree has the least standard deviation versus the rest of these levels of education.

___

<a id='plot5'></a>

__`sns.barplot(data=, x='', y='', estimator=, ci=, hue='')`__

```python
# the average salary per level of education per division

plt.figure(figsize=(9, 5))
sns.barplot(data=df, x='level of education', y='salary', estimator=np.mean,
           ci='sd', hue='division')
```

<a id='plot6'></a>

__`plt.legend(bbox_to_anchor=())`__

```python
plt.figure(figsize=(9, 5), dpi=80)
sns.barplot(data=df, x='level of education', y='salary', estimator=np.mean,
           ci='sd', hue='division')

plt.legend(bbox_to_anchor=(1.05, 1))
```
