grid - сетка

- seaborn grids are essentially analogous to matplotlib subplots call, and what they do is they automatically create a grid for these subplots based off a categorical column.
- for the matplotlib subplots we have to pass in a specific number of columns or rows, but with seaborn, what we can do is we can actually just pass in the name of the column and/or row that we want and seaborn will automatically map to the subplots grid.
- directly calling the grid system allows users to heavily customize plots.

## Catplot

So this is essentially the statistical distributions within categories. But more importantly, it takes in these additional arguments of _row_ and _column_ which is going to define the subplot grid. And then we just have to say what kind of plot we want.

So if I run this, this looks like a normal boxplot call, except in this case, it's the more generalized categorical plot call with kind specified to box.

__`sns.catplot(data=, x='', y='', kind='box')`__

```python
sns.catplot(data=df,
           x='gender',
           y='math score',
           kind='box')
```

---
What's so special about this is that then I have these extra parameters I can call, which is row and column.

Now it's going to create rows for every category of lunch.

__`sns.catplot(data=, x='', y='', kind='box', row='')`__

```python
sns.catplot(data=df,
           x='gender',
           y='math score',
           kind='box',
           row='lunch')
```

So here each row for the subplot grid represents a different lunch type.

This is really similar to setting hue='lunch'.

__`sns.catplot(data=, x='', y='', kind='box', hue='')`__

## PairGrid

Grid that pairplot is built on top of, allows for heavy customization of the pairplot seen earlier.

__`sns.PairGrid(data=)`__

```python
sns.PairGrid(df)
```

At first it just creates the grid automatically for us based off the columns. 

---
What it is still waiting for are the actual mapping commands that it wants us to provide.

__`g = sns.PairGrid(data=)`__<br>__`g = g.map_upper(sns.scatterplot)`__

```python
g = sns.PairGrid(df)
g = g.map_upper(sns.scatterplot)
```

---

__`g = sns.PairGrid(data=)`__<br>__`g = g.map_lower(sns.kdeplot)`__

```python
g = sns.PairGrid(df)
g = g.map_lower(sns.kdeplot)
```

---

__`g = sns.PairGrid(data=)`__ <br> __`g = g.map_diag(sns.kdeplot)`__

```python
g = sns.PairGrid(df)
g = g.map_diag(sns.kdeplot)
```

---

__`g = sns.PairGrid(data=)`__<br>__`g = g.map_upper(sns.scatterplot)`__
<br>__`g = g.map_lower(sns.kdeplot)`__<br>__`g = g.map_diag(sns.kdeplot)`__

```python
g = sns.PairGrid(df)
g = g.map_upper(sns.scatterplot)
g = g.map_diag(sns.kdeplot)
g = g.map_lower(sns.kdeplot)
```

---

__`g = sns.PairGrid(data=)`__<br>__`g = g.map_upper(sns.scatterplot, alpha=)`__
<br>__`g = g.map_lower(sns.kdeplot, color='')`__<br>__`g = g.map_diag(sns.histplot)`__

```python
g = sns.PairGrid(df)
g = g.map_upper(sns.scatterplot, alpha=.2)
g = g.map_diag(sns.histplot)
g = g.map_lower(sns.kdeplot, color='red')
```

---

Let's say I want to add in a hue setting for this entire plot. I can add in those higher level parameters.

__`g = sns.PairGrid(data=, hue='')`__<br>__`g = g.map_upper(sns.scatterplot)`__
<br>__`g = g.map_lower(sns.kdeplot)`__<br>__`g = g.map_diag(sns.histplot)`__

```python
g = sns.PairGrid(df, hue='gender')
g = g.map_upper(sns.scatterplot)
g = g.map_diag(sns.histplot)
g = g.map_lower(sns.kdeplot)
```

---

One thing to notice here is, for whatever reason, it doesn't automatically add the legend. So for that, you have to actually add in the legend manually.

__`g = sns.PairGrid(data=, hue='')`__<br>__`g = g.map_upper(sns.scatterplot)`__
<br>__`g = g.map_lower(sns.kdeplot)`__<br>__`g = g.map_diag(sns.histplot)`__<br>__`g = g.add_legend()`__

```python
g = sns.PairGrid(df, hue='gender')
g = g.map_upper(sns.scatterplot)
g = g.map_diag(sns.histplot)
g = g.map_lower(sns.kdeplot)
g = g.add_legend()
```
