<u>Distribution within categories</u>
- Boxplot
- Violinplot
- Swarmplot
- Boxenplot (Letter-Value Plot)

___

- __Boxplot__ displays the distribution of a continuous variable.
- It does this through the use of quartiles.
- 25 percent of the data points are below Q1.
- 50th percentile (Q2) is the median.
- Q3 is the 75th percentile.
- IQR (Interquartile Range) defines the box width and it's essentially the middle 50 percent of all the data points, between Q1 and Q3.
- The 'whiskers' are defined by 1.5 x IQR
- Outside of the whiskers are outliers.

```python
from PIL import Image 
Image.open('photo_2022-05-10_05-11-43.jpg')
```

___

- __Violinplot__ displays the probability density across the data using a KDE.
- We can imagine it as a mirrored KDE plot. (it's the KDE plot being mirrored and joined to itself)
- A little white dot in the middle is the median. 
- A box within the violin plot is the interquartile range for the violin plot.

___

- __Swarmplot__ just shows all the data points in the distribution. 
- Keep in mind, if you have a very large data set, it's actually not going to be able to show you all the points. And instead, seaborn is going to warn you that you're missing a certain percentage of points. It won't show all the points, but will display the general distribution of them.
- swarmplot is not a mirror image.

___

- __Boxenplot (Letter-value plot)__ is mainly designed as an expansion upon the normal boxplot.
- It essentially uses multiple quantiles instead of strictly quartiles.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
df = pd.read_csv('StudentsPerformance.csv')
```

```python
df.head()
```

```python
df.info()
```

___

## Boxplot

The following boxplot displays distribution for all of the math scores not separated by any category.

```python
sns.boxplot(data=df, y='math score')
```

If we actually do want to separate it by category, then you must specify both x and y.

```python
sns.boxplot(data=df,
            y='math score', 
            x='test preparation course')
```

```python
plt.figure(figsize=(8, 5))
sns.boxplot(data=df, 
            x='math score', 
            y='test preparation course')
```

___

```python
plt.figure(figsize=(10, 5), dpi=80)
sns.boxplot(data=df, 
            x='parental level of education',
            y='reading score')

# So now we can see how the parental level of education affects test scores such as the reading score.
```

We could then do subcategories based off another categorical feature, simply by specifying a `hue` value, which essentially creates subcategories within your larger categories.

```python
plt.figure(figsize=(10, 5), dpi=80)
sns.boxplot(data=df, 
            x='parental level of education', 
            y='reading score',
            hue='test preparation course')

plt.legend(bbox_to_anchor=(1.05, 0.15))
```

```python
# the same just changing the axes

plt.figure(figsize=(6, 6), dpi=80)
sns.boxplot(data=df, 
            y='parental level of education',
            x='reading score',
            hue='test preparation course',
            palette='Set2')

plt.legend(bbox_to_anchor=(1.05, 0.15))
```

___

## Violinplot

```python
plt.figure(figsize=(9.8, 5))

sns.violinplot(data=df,
              x='parental level of education',
              y='reading score')
```

```python
plt.figure(figsize=(7, 7))

sns.violinplot(data=df,
              x='reading score',
              y='parental level of education')
```

```python
plt.figure(figsize=(7, 7))

sns.violinplot(data=df,
              x='reading score',
              y='parental level of education',
              hue='test preparation course',
              palette='Set3')

plt.legend(bbox_to_anchor=(1.05, .5))
```

What `split` does is instead of stacking these together, it makes one violin based off the top half of one category and the bottom half of another.

```python
plt.figure(figsize=(7, 7))

sns.violinplot(data=df,
              x='reading score',
              y='parental level of education',
              hue='test preparation course',
              palette='Set3', 
              split=True)

plt.legend(bbox_to_anchor=(1.05, .5))
```

To get rid of mini boxplot in the middle representing those quartiles and white dot for the median, set `inner=None`

```python
plt.figure(figsize=(7, 7))

sns.violinplot(data=df,
              x='reading score',
              y='parental level of education',
              hue='test preparation course',
              palette='Set3',
              split=True,
              inner=None)

plt.legend(bbox_to_anchor=(1.3, .5))
```

To see the quartiles for each KDE drawn  `inner='quartile'`

```python
plt.figure(figsize=(7, 7))

sns.violinplot(data=df,
              x='reading score',
              y='parental level of education',
              inner='quartile')
```

`inner=stick` is essentially going to draw a line for every instance, kind of showing you the distribution along with the KDE.

```python
plt.figure(figsize=(7, 7))

sns.violinplot(data=df,
              x='reading score',
              y='parental level of education',
              inner='stick')
```

When we first learned about the KDE plot, there was something called the `bandwidth parameter`, essentially how much noise you want to pick up when you're stacking and summing together those Gaussian distributions.

So if I were to set bandwidth to something smaller, like 0.1, then these KDEs are going to start picking up more noise.

```python
plt.figure(figsize=(7, 7))

sns.violinplot(data=df,
              x='reading score',
              y='parental level of education',
              bw=0.1)
```

```python
# zero point zero one is going to be extremely noisy and definitely not useful

plt.figure(figsize=(7, 7))

sns.violinplot(data=df,
              x='reading score',
              y='parental level of education',
              bw=0.01)
```

## Swarmplot

```python
sns.swarmplot(data=df,
             x='math score')
```

It gives you a warning and it says 15.8% of the points cannot be placed. So essentially the issue here is this figure size or the dot size are not the right size in order to display all the points.

To change the dot size, use `size` parameter.

```python
sns.swarmplot(data=df,
             x='math score', 
             size=2)
```

And just as before, if you want, you can separate this out based off some categorical column.

```python
plt.figure(figsize=(8, 4), dpi=80)

sns.swarmplot(data=df,
             x='math score',
             y='gender',
             size=2)
```

```python
plt.figure(figsize=(8, 4), dpi=80)

sns.swarmplot(data=df, 
             x='math score',
             y='gender',
             hue='test preparation course',
             size=2)

plt.legend(bbox_to_anchor=(1.25, .4))
```

The above figure is really difficult to read, so it would be useful to use `dodge=True` parameter, which is going to split up these swarmplots per that secondary hue category that you asked for.

```python
plt.figure(figsize=(8, 4), dpi=80)

sns.swarmplot(data=df, 
             x='math score',
             y='gender',
             hue='test preparation course',
             dodge=True,
             size=2)

plt.legend(bbox_to_anchor=(1.25, .4))
```
