```python
import pandas as pd
```

What pandas does is it just grabs all the tables from a page and then you can go ahead and explore what tables you're interested in.

```python
url = 'https://en.wikipedia.org/wiki/World_population'
```

```python
tables = pd.read_html(url)

# It always returns a list. Even if there's just one table, it's going to return a list of one item.
```

You should also __keep in mind__ that sometimes working on a company Internet, your firewall may be blocking you from doing this sort of functionality. And in that case what you can do is right click on this -> view page source, which shows you the entire HTML document -> right click and save this as an HTML file -> pass the HTML file instead of the URL in pd.read_html()

__`save or export to an HTML file`__

```python
world_topten = tables[0]
world_topten
```

```python
world_topten = world_topten.drop(11, axis=0)
world_topten = world_topten.drop('#', axis=1)
world_topten
```

```python
world_topten.to_html('sample_table.html', index=False)
```
