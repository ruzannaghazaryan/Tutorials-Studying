- How can we improve (adjust) on a Linear Regression Model?
- One approach is to consider __higher order relationships__ on the features.

---

There are two issues that polynomial regression is going to address for us.
- __Non-linear feature relationships to label__
- __Interaction terms between the features themselves__

## Non-linear feature relationships to label

```python
from PIL import Image
```

```python
Image.open('nonlinear.png')
```

Let's imagine we have a feature that is non linear, so we have some sort of x feature and that x feature actually behaves a little more like something like log of x. __Keep in mind__, the log of X here, that x inside is not the feature itself. What we're here just saying is that we have some sort of feature that behaves like log of x, x inside of log is not the actual feature.

So we know log of x is not a linear relationship. __So what do we do if we have a feature x that behaves similar to log of x?__

Well, unfortunately, it's going to be difficult if we just apply linear regression to a feature like this to actually find a beta coefficient that accurately grabs information from this feature because it's difficult to find a linear relationship on this logarithmic curve.

```python
Image.open('linnon.png')
```

But what about the square of this feature? So if we actually take the square of this, then notice it starts to become more linear. Then we could even go to higher orders such as the third power. And you'll notice it really starts to appear linear here compared to those original curves, which makes it much easier to find a good beta coefficient for those higher orders than the original feature.

```python
Image.open('more.png')
```

So, that's essentially the motivation behind polynomial regression is that you will be able to find those accurate beta coefficients for higher orders of the feature if the relationship between the original feature is non linear.

- Now, __keep in mind__, we basically chose a very convenient example where your feature behaves exactly like log of x. Not every feature is going to have relationships at a higher order.
- The main point here is to show that in certain situations it could be reasonable to solve for a single linear beta coefficient for polynomial of an original feature.

## Interaction terms

__Interaction terms__ just means that __what happens if two features are only significant when they're in sync with one another?__ And this is sometimes known in the business world as __synergy__.

So, for example, perhaps newspaper advertising spend by itself is not effective, but it greatly increases the effectiveness if added to an additional TV advertising campaign. For example, consumers only watching a TV advertisement will create some sales, but consumers who maybe watch a TV advertisement and are later reminded of this through an additional newspaper advertisement could actually contribute even more sales than if you had only used TV or newspaper on their own.

So how can we quantitatively check if there is a synergy between multiple features? In this case, the simplest way to do that is to simply create a new feature that multiplies two existing features together.

We can also keep the original features. I can create the feature of the interaction term while keeping x1 and x2.

---

So polynomial regression is going to create the following.
- It will create the bias, basically a default value of 1 indicating the y intercept term.
- Then it's going to create values raised to a power for each degree.
- Then it also creates interaction features between all pairs of possible feature combinations.

---

__Keep in mind__ there's no guarantee that there are actually signals within interaction terms or higher order relationships, but you won't really know until you try.
