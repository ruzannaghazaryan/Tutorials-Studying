### Columns

<u>grabbing a __single column__

```python
df['total_bill']
```

```python
type(df['total_bill'])
```

What if I want to __grab multiple columns__? In that case, I pass __a list of column names__.

```python
df[['total_bill', 'tip']]
```

Let's imagine I want to have a column that told me the tip percentage, essentially, what is the percentage of the tip in accordance to the total bill?

__`creating a new column`__

```python
df['tip_percentage'] = df['tip'] / df['total_bill'] * 100
df['tip_percentage']
```

```python
df.head(2)
```

___

Our tip percentage has a lot of floating point values here, maybe that's not something we want to do.

__So because pandas is based on numpy, we get the really nice capabilities of using numpy universal functions along with pandas.__

```python
df['tip_percentage'] = np.round(df['tip'] / df['total_bill'] * 100, 2)
```

```python
df.head(2)
```

___

__`removing columns`__

```python
df.drop('tip_percentage')
```

That's because the default axis is axis equal to zero (meaning rows).
<br>So it's looking for a labeled index called tip percentage.

```python
df.drop('tip_percentage', axis=1)
```

```python
df.head(3) # it still has a tip percentage column
```

## !!!!!

__A really important thing to note here is that pandas, in general, when you call operations like drop or just different method, typically does NOT do them in place. That is to say, it does not permanently affect your data frame.__

<u>If I want this change to be permanent. There's two options here.

- we can see if I do shift tab in .drop(), set inplace equal to True

```python
df.drop('tip_percentage', axis=1, inplace=True)
```

```python
df.head()
```

- or just say df is equal to what this actually returns, which is the dataframe with that column dropped.

__this one is recommended to do__

```python
df['tip_percentage'] = df['tip'] / df['total_bill'] * 100
```

```python
df = df.drop('tip_percentage', axis=1)
df.head(3)
```
