The plt.subplots() command returns a tuple containing the Figure canvas and then a numpy array holding the axes objects.

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
fig, axes = plt.subplots(nrows=1, ncols=2)
```

___

```python
# exponential relationship

a = np.linspace(0, 10, 11)
b = a ** 4 # (a to the power of 4)
```

```python
# linear relationship

x = np.arange(0, 10)
y = 2 * x
```

__!!! Keep in mind__, everything should be going in the same cell.

```python
fig, axes = plt.subplots()

axes.plot(x, y)
```

By default it just creates one row by one column, which means the axis is actually just a single set of axes.

```python
axes
```

<a id='id2'></a>

```python
fig, axes = plt.subplots()
axes.plot(x, y, 'r')
axes.set_title('Title')
axes.set_xlabel('x')
axes.set_ylabel('y');
```

So because this is now essentially a single dimensional array, you can actually loop through it.

```python
fig, axes = plt.subplots(nrows=3, ncols=1)

for ax in axes:
    ax.plot(x, y)
```

<a id='id7'></a>
We have the issue of everything seems to be very close together. 

1) use __plt.tight_layout()__  at the very bottom of the same cell.
<br>And what it's going to do is it's going to automatically attempt to make sure that no set of axes' tickers are overlapping over anything else.

2) use __.subplots_adjust(wspace=, hspace=) method__ to manually set the spacing.
https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.subplots_adjust.html

```python
fig, axes = plt.subplots(nrows=3, ncols=1)

axes[0].plot(x, y)

plt.tight_layout()
```

wspace, hspace _ the way it's expressed is as a fraction of the average axis width, height.

```python
fig, axes = plt.subplots(nrows=2, ncols=2)

axes[0][0].plot(x, y)

axes[0][1].plot(x, y)

axes[1][0].plot(x, y)

axes[1][1].plot(a, b)

fig.subplots_adjust(wspace=.9)
```

```python
fig, axes = plt.subplots(nrows=2, ncols=2)

axes[0][0].plot(x, y)

axes[0][1].plot(x, y)

axes[1][0].plot(x, y)

axes[1][1].plot(a, b)

fig.subplots_adjust(wspace=1, hspace=1)
```

___

<a id='id8'></a>
### editing the parameters of the subplots

```python
fig, axes = plt.subplots(nrows=2, ncols=2)

axes[0][0].plot(x, y)

axes[0][1].plot(x, y)

axes[1][0].plot(x, y)
axes[1][0].set_ylabel('Y label 1, 0')
axes[1][0].set_xlim(0, 4)

axes[1][1].plot(a, b)
axes[1][1].set_title('Title 1, 1')

plt.tight_layout()
```

___

<a id='id9'></a>
__A title on the entire plot__

```python
fig, axes = plt.subplots(nrows=2, ncols=2)

axes[0][0].plot(x, y)

axes[0][1].plot(x, y)

axes[1][0].plot(x, y)
axes[1][0].set_ylabel('Y label 1, 0')
axes[1][0].set_xlim(0, 4)

axes[1][1].plot(a, b)
axes[1][1].set_title('Title 1, 1')

fig.suptitle('Figure Level Title', fontsize=16) # super title

plt.tight_layout()
```

___

<a id='id11'></a>
### saving the figure

```python
fig, axes = plt.subplots(nrows=2, ncols=2)

axes[0][0].plot(x, y)

axes[0][1].plot(x, y)

axes[1][0].plot(x, y)
axes[1][0].set_ylabel('Y label 1, 0')
axes[1][0].set_xlim(0, 4)

axes[1][1].plot(a, b)
axes[1][1].set_title('Title 1, 1')

fig.suptitle('Figure Level Title', fontsize=16) # super title

plt.tight_layout()

fig.savefig('studying_subplots.png', bbox_inches='tight') # bounding box inches
```
