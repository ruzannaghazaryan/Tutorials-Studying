- __A relative path__ will get you there from where you are now. If you use it from somewhere else, you won't get where you want to be.

## `Reading from a text file`

__The basics of reading from a text file__

The steps are: 
- Open the file
- Read the data
- Close the file

After reading the data, you'd either process it in some way, or store it somewhere.

We want to read from the file, so we open it for reading. To open a file, we use the open function.

```python
jabber = open('Jabberwocky.txt', 'r')

# Because I put the text file into my project directory, I haven't had to specify the full path.

# The second argument is 'r', which stands for read. We can read from this file, but we can't write to it. 
```

This isn't the cleanest way to open a file, in Python. PEP 343 introduced the <u>_with_</u> statement, for working with things like files. So for now, don't take this as the best way to open a file – think of it as a way to understand how files work.

The __open() function__ returns an object that we can use to get data from the file. In this example, we bind the object to a variable called jabber.

Text files contain lines of text, and we can iterate over jabber, to get each line:

```python
for line in jabber:
    print(line)
    
jabber.close()
```

The for loop binds each line from the file, to the variable line, then prints it out.
<br>__Importantly__, we have to __close the file__ when we're finished. If you don't close a file when you're finished with it, you can get problems.

Why is our output double spaced? The reason for that is, there's a newline character at the end of each line, in the file. (In Python strings \n). When text editors see a newline in the file, they don't print it – they start a new line instead. So although you can't see it, there's an extra character at the end of each line. You can't see the character, you see its behaviour instead – it causes a new line to be started. The print function defaults to printing \n when it prints. So we have a newline printed because each line of text ends with a newline, and we get another newline printed automatically, by the print function.

```python
jabber = open('Jabberwocky.txt', 'r')

for line in jabber:
    print(line, end='')
    
jabber.close()
```

## Opening a file using `with`

Let's do the same thing in a more Pythonic way:

```python
with open('Jabberwocky.txt', 'r') as jabber:
    for line in jabber:
        print(line.rstrip())
```

Line 1 uses the open function, just like before. But we use a _with_ statement. Because with is a statement, not a function, we don't get a file object returned. Instead, we use "as jabber" to give a name to the file object.

#### *** Opening a file this way is preferred, because the file will be automatically closed when the with block terminates. There's no chance that we'll forget to close the file.

---

### `.readline() method`

```python
with open('Jabberwocky.txt') as jabber:
    while True:
        line = jabber.readline().rstrip()
        print(line)
        if 'jubjub' in line.casefold():
            break
```

Here, we read the file line-by-line, and break out of the loop when we find something we're looking for. It reads one line of the file at a time. The reason you might not use it very often, is that we can get the same result by just iterating over the file object – which is what we did when we first looked at reading from a text file.

```python
# Here is the equivalent code:

with open('Jabberwocky.txt') as jabber:
    for line in jabber:
        print(line.rstrip())
        if 'jubjub' in line.casefold():
            break
```

They both behave very similarly. You may want to use the first loop – with readline – if you want to process the string in some way, and work with the processed string throughout the remainder of the code. For example, on line 3, we strip the newline from the end of the string, before binding the variable line to it. Compare that to line 5, where we also strip the newline, but only for printing. The test, on line 6, is performed with a string that still contains the newline. Strings are immutable, remember. So line 5 doesn't mutate the string – it just prints a copy that doesn't end with a newline.

__The default mode__, if you don't specify __the second argument in open() function__, is to open the file for reading.
So that's why I left __"r"__ out of those three function calls.

### `.startswith(), .endswith() methods`

It's quite easy to create functions that perform the same job as these new methods.

```python
def removeprefix(string: str, prefix: str) -> str:
    if string.startswith(prefix):
        return string[len(prefix):]
    else:
        return string[:] # return a copy of `string`
```

```python
def removesuffix(string: str, suffix: str) -> str:
    if suffix and string.endswith(suffix): # the right code
        return string[:-len(suffix)]
    else:
        return string[:]
```

```python
first
```

```python
removeprefix(first, "'Twas")
```

```python
removesuffix(first, 'toves')
```

```python
removesuffix(first, '')
```

```python
def removesuffix(string: str, suffix: str) -> str:
    if string.endswith(suffix): # difference is here
        return string[:-len(suffix)]
    else:
        return string[:]
```

```python
removesuffix(first, '')
```

pass an empty string as the suffix. You'll get an empty string returned because len("") is zero, and string[:0] is an empty string.

This happens because all strings return True for .endswith("") - the empty string is effectively in all strings, including at their end.

---

return string[:]

The function is described as returning a new string, so we ensure that the behaviour matches the description. Returning a new string in some cases, and the original string in others, could result in subtle bugs when the function is used.

## `Parsing`

In this case, the fields are separated by a pipe character.

It's essential that __each row has the same number of fields__.
<br>__When a piece of data is missing, it's essential that the separator is still present.__

Making sense of data is called __parsing__. When we talk about __parsing data__ – or __parsing code__ – it means __splitting it up into logical components__.

Let's get Python to parse this country data.

```python
input_filename = 'country_info.txt'
```

Next, we'll open the file, and iterate over the lines of text:

```python
with open(input_filename) as country_file:
    for row in country_file:
        # We've seen that each field is separated from the next by a pipe character.
        data = row.split('|')
        print(data)
```

There's a minor problem – the last field ends with a newline character. We can fix that – as we've seen – using the strip method.

```python
with open(input_filename) as country_file:
    for row in country_file:
        data = row.strip('\n').split('|')
        print(data)
```

We read the file, and created lists for each country, containing the seven pieces of information for each one.

```python
with open(input_filename) as country_file:
    for row in country_file:
        data = row.strip('\n').split('|')
        country, capital, code, code3, dialing, timezone, currency = data
        print(country, capital, code, code3, dialing, timezone, currency, sep='\n\t')
```

We can unpack the list for each country. There are seven fields, so we'll use seven variables.
This is why it's essential that every line in the file contains seven fields –
even if some fields are blank. If a list had fewer than seven items,
our code would crash when we tried to unpack it into seven variables.

Right at the top of the output, the first row in the file contained the headings for the fields.
That can be useful, but we don't want to include those entries in our dictionary.

The simplest way to exclude them, is to read that first line, before we iterate over the remainder of the file. I'll add a call to the readline method, before the loop. Our country_file object is iterable. Each time you use the readline method to read a line, the iterator automatically moves on to the next item. That's the next line, in this case. So our for loop, on line 3, starts with the second line of the file – because we've already read the first line, on line 2.

```python
with open(input_filename) as country_file:
    country_file.readline()
    for row in country_file:
        data = row.strip('\n').split('|')
        country, capital, code, code3, dialing, timezone, currency = data
        print(country, capital, code, code3, dialing, timezone, currency, sep='\n\t')
```

```python
with open(input_filename) as country_file:
    country_file.readline()
    for row in country_file:
        data = row.strip('\n').split('|')
        country, capital, code, code3, dialing, timezone, currency = data
        country_dict = {
            'name': country,
            'capital': capital,
            'country_code': code,
            'cc3': code3,
            'dialing_code': dialing,
            'timezone': timezone,
            'currency': currency
        }
        print(country_dict)
```

What we need now, is some way to choose the country.
We'll start by choosing a country by name. That means we need to put each of these
dictionaries in another dictionary, with the key being the country name.

```python
countries = {}
with open(input_filename) as country_file:
    country_file.readline()
    for row in country_file:
        data = row.strip('\n').split('|')
        country, capital, code, code3, dialing, timezone, currency = data
        country_dict = {
            'name': country,
            'capital': capital,
            'country_code': code,
            'cc3': code3,
            'dialing_code': dialing,
            'timezone': timezone,
            'currency': currency
        }
        countries[country.casefold()] = country_dict
print(countries)
```

## `Writing data to a text file`

```python
plants_filename = 'flowers_write.txt'

with open(plants_filename, 'w') as plants:
    for plant in data:
        plants.write(plant)
```

Instead of each plant on a new line, we've got a single line of text, with each plant written immediately after the previous one.
That's the most obvious difference between using print and using write.



__Remember__ that text files can only hold text.
If you try to put numbers into a text file, things don't work.
But you can put the string representation of a number into a text file.
That means we can print numbers to a file, but we can't write them.

```python
filename = 'test_numbers.txt'

with open(filename, 'w') as test:
    for i in range(10):
        print(i, file=test)
```

Because print automatically adds a newline, each number is on a new line. 

If we do the same thing with the write method, it doesn't work:

```python
with open(filename, 'w') as test:
    for i in range(10):
        test.write(i)
```

We're writing to a text file, which means we can only put strings into it.

---

The __print()__ function will print the string representation of any object that you ask it to print. 

In addition, it will include a separator between multiple arguments – the default is a space, but that can be changed with the sep keyword argument.

Finally, it will print the value of the end keyword argument. The default is a newline character.



---

The __.write()__ method will only write exactly what you tell it to write.

No separators or newline characters are included, unless you explicitly tell it to write them.

Also, no conversion is performed. If you tell write to write an integer, that's what it will try to send to the file.

If the file is opened in text mode (the default), you'll get an error if you try to write numerical values to it.

---

So which should I use, write or print? The answer to that is, use whichever is more convenient for the task you're trying to perform. 

If you want complete control over what gets written, use write.

If it's more convenient to let print include separators and newline, and convert objects to their string representations, then use print.

---

```python
filename = 'test_numbers_w.txt'

with open(filename, 'w') as test:
    for i in range(10):
        test.write(str(i))
```

The .write() method writes exactly what we tell it to write – nothing less, nothing more.

```python
filename = 'test_numbers_w_newline.txt'

with open(filename, 'w') as test:
    for i in range(10):
        test.write(str(i) + '\n')
```

## `File Modes`

https://docs.python.org/3/library/functions.html#open

- __mode "w"__ will create a file, if it doesn't already exist, and clear it out, if it does exist.<br>
In either case – whether the file already existed or not – you start off with an empty file.
- __mode "x"__ will create a new file, but will raise an error if the file already exists. Mode "x" will guard against accidentally deleting the contents of an existing file. But you'll need to catch the error, if you don't want your code to crash, when the file does exist.
- If you want to add more text to the end of an existing file, you'd specify __"a"__ for the __mode__. That will create a new file if one doesn't already exist, and appends to the end of the existing file, if there is one. It's similar to mode "w", but doesn't truncate the file.

## `The CSV format`

__CSV__ stands for __Comma Separated Values__. The basic idea is that each value on a row is separated from the preceding and following values, with a comma. A CSV file contains rows of data, with each row having the same number of "fields". Each field is separated by some character. That can be a comma (hence the name), but you can also use tabs, or spaces, or any other character that's appropriate.

---

```python
with open(csv_filename, encoding='utf-8', newline='') as csv_file:
    reader = csv.reader(csv_file)
    for row in reader:
        print(row)
```

All the numerical values have been read as strings. That's what we're going to fix.

All the strings that really are strings have been quoted correctly. If we tell the reader function that they are quoted, the reader function can take care of converting the numerical values to their correct types.

```python
with open(csv_filename, encoding='utf-8', newline='') as csv_file:
    reader = csv.reader(csv_file, quoting=csv.QUOTE_NONNUMERIC)
    for row in reader:
        print(row)
```

Now, we've got floats rather than strings. That was easy! But it does have some limitations. The main limitation is that it only works with data that has all string values quoted. The other problem is that the numerical values are all floats. That's probably fine, most of the time, but is something to be aware of. If you need integer values to be converted to ints (not floats), then you can't use this approach.

### `Sniffer and Dialect`

This code doesn't specify the field separator character, so it's probably not going to work. Run the program, to check. We get lists containing a single string – the contents of the entire line, for each row in the file. If we want the reader to parse this data, we have to tell it how the fields are separated.

```python
import csv

input_filename = 'country_info.txt'

with open(input_filename, encoding='utf-8', newline='') as countries_data:
    country_reader = csv.reader(countries_data)
    for row in country_reader:
        print(row)
```

```python
with open(input_filename, encoding='utf-8', newline='') as countries_data:
    country_reader = csv.reader(countries_data, delimiter='|')
    for row in country_reader:
        print(row)
```

The CSV module can make life easier, by checking the data for you. To do that, we use the CSV Sniffer class. It examines a sample of the file, and works out things like the separators, the character used to delimit strings, and so on. It uses that sample to create a Dialect.

Run the program and see it won't work.

```python
with open(input_filename, encoding='utf-8', newline='') as countries_data:
    sample = countries_data.read()
    country_dialect = csv.Sniffer().sniff(sample)
    country_reader = csv.reader(countries_data, dialect=country_dialect)
    for row in country_reader:
        print(row)
```

Didn't work. We don't get any output. On line 2, we read the entire contents of the file. Because we've read the entire file, there's nothing left for the reader to read. To fix the problem, we need to tell Python to go back to the start of the file: The .seek() method is used to position the file pointer to another place in the file.

```python
with open(input_filename, encoding='utf-8', newline='') as countries_data:
    sample = countries_data.read()
    country_dialect = csv.Sniffer().sniff(sample)
    countries_data.seek(0)
    country_reader = csv.reader(countries_data, dialect=country_dialect)
    for row in country_reader:
        print(row)
```

The Sniffer worked out that the fields are delimited with a pipe character, and the reader used the dialect that sniff created, to correctly parse the data.

At the moment, we read the entire file into a string, to sniff it. We then reset the file pointer to the start of the file – that's the code on line 4 – and the entire file gets read again, by the reader. That's obviously not very efficient, especially with very large files. The sniffer only needs a sample of the data. Generally, two or three lines should be sufficient for it to work out the format.

I'll read 3 lines from the file, and use them to produce the sample string.

```python
with open(input_filename, encoding='utf-8', newline='') as countries_data:
    sample = ''
    for line in range(3):
        sample += countries_data.readline()
    country_dialect = csv.Sniffer().sniff(sample)
    countries_data.seek(0)
    country_reader = csv.reader(countries_data, dialect=country_dialect)
    for row in country_reader:
        print(row)
```
